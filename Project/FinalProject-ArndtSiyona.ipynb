```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
print(sns.__version__)
```

```python
# Open the CSV file
Diamonds = pd.read_csv("Diamonds.csv")

# Create the seaborn plot
sns.set_palette('colorblind')
fig1 = sns.displot(Diamonds, x = 'price', hue = 'cut', bins = 50)
fig1.set(title = 'Seaborn - Price of diamonds based on cut type', xlabel = 'price [dollars]')

# Separate out the data to create a matploptlib style plot
ideal = Diamonds[Diamonds['cut'] == 'Ideal']
premium = Diamonds[Diamonds['cut'] == 'Premium']
good = Diamonds[Diamonds['cut'] == 'Good']
very_good = Diamonds[Diamonds['cut'] == 'Very Good']
fair = Diamonds[Diamonds['cut'] == 'Fair']

# Create matplotlib style plot
bins = 50
alpha = .7
fig2, ax2 = plt.subplots()
ax2.hist(ideal['price'], bins = bins, alpha = alpha, label = 'Ideal', color = 'lightblue')
ax2.hist(premium['price'], bins = bins, alpha = alpha, label = 'Premium', color = 'yellow')
ax2.hist(fair['price'], bins = bins, alpha = alpha, label = 'Fair', color = 'purple')
ax2.hist(very_good['price'], bins = bins, alpha = alpha, label = 'Very Good', color = 'orange')
ax2.hist(good['price'], bins = bins, alpha = alpha, label = 'Good', color = 'lightgreen')
ax2.set_title('Matplotlib - Price of diamonds based on cut type')
ax2.set_ylabel('Counts')
ax2.set_xlabel('price [dollars]')
ax2.legend(); 
```

The seaborn plot is on top and the matplotlib plot is on the bottom. The matplotlib plot took vastly more code to plot the differnt cuts on top of eachother, but changing individual aspects of the plot is easier.

```python
# Create the seaborn plot for a 2D hist 
ax3 = sns.displot(Diamonds, x = 'carat', y = 'price', hue = 'cut', binwidth = (0.1, 250), cbar = True)
```

This is what the initial formating looks like for a 2D histogram sorted by the 5 different data types. It is not clean and it is not readable

```python
ax4 = sns.displot(Diamonds, x = 'carat', y = 'price', hue = 'cut', 
                  binwidth = (0.1, 250), cbar = True, aspect = 3)
ax4.figure.set_size_inches(12, 5)
sns.move_legend(ax4, 'lower center', bbox_to_anchor = (0.3, 0.12))
```

This is a better layout than it was origionally, however the ratio of the size of the plot to the color bars is not ideal.

```python
# Create a violin plot very fast shwoing mean, upper and lower quartiles, and maxes and mins.

fig5 = sns.catplot(Diamonds, x = 'cut', y = 'price', kind = 'violin')
fig5.set(title = 'Price of diamonds based on cut type',
        ylabel = 'price [dolars]')
```

This is for plotting categroical data. This works well in this case and is very fast and the plot looks nice.

```python
# Creating a scatter plot of data
ax5 = sns.relplot(data = Diamonds, x = 'depth', y ='price', col = 'cut', col_wrap = 2)
```

This shows graphs of the depth versus the price of diamonds, with a different plot created for each cut style. Again this was able to be created with very little code, and looks looks readable. However, creating a more polished plot takes more time and code.
