```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
print(sns.__version__)
```

```python
# Open the CSV file
Diamonds = pd.read_csv("Diamonds.csv")

# Create the seaborn plot
sns.set_palette('colorblind')
fig1 = sns.displot(Diamonds, x = 'price', hue = 'cut', bins = 50)
fig1.set(title = 'Seaborn - Price of diamonds based on cut type', xlabel = 'price [dollars]')

# Separate out the data to create a matploptlib style plot
ideal = Diamonds[Diamonds['cut'] == 'Ideal']
premium = Diamonds[Diamonds['cut'] == 'Premium']
good = Diamonds[Diamonds['cut'] == 'Good']
very_good = Diamonds[Diamonds['cut'] == 'Very Good']
fair = Diamonds[Diamonds['cut'] == 'Fair']

# Create matplotlib style plot
bins = 50
alpha = .7
fig2, ax2 = plt.subplots()
ax2.hist(ideal['price'], bins = bins, alpha = alpha, label = 'Ideal', color = 'lightblue')
ax2.hist(premium['price'], bins = bins, alpha = alpha, label = 'Premium', color = 'yellow')
ax2.hist(fair['price'], bins = bins, alpha = alpha, label = 'Fair', color = 'purple')
ax2.hist(very_good['price'], bins = bins, alpha = alpha, label = 'Very Good', color = 'orange')
ax2.hist(good['price'], bins = bins, alpha = alpha, label = 'Good', color = 'lightgreen')
ax2.set_title('Matplotlib - Price of diamonds based on cut type')
ax2.set_ylabel('Counts')
ax2.set_xlabel('price [dollars]')
ax2.legend(); 
```

```python
# Create the seaborn plot for a 2D hist 
# This is what the inital fomrating looks like for 5 differnt categories. 
# It is not clean and it is not readable
ax3 = sns.displot(Diamonds, x = 'carat', y = 'price', hue = 'cut', cbar = True)
ax3.set(title = 'Seaborn (inital format) - Price of diamonds based on cut type and carat',
        ylabel = 'price [dolars]')
```

```python
# Create a violin plot very fast shwoing mean, upper and lower quartiles, and maxes and mins.
# This is for plotting categroical data
# This works well in this case and is very fast and the plot looks nice

fig4 = sns.catplot(Diamonds, x = 'cut', y = 'price', kind = 'violin')
fig4.set(title = 'Price of diamonds based on cut type',
        ylabel = 'price [dolars]')
```
